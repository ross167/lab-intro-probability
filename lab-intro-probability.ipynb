**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
# Challenge ones

from scipy.stats import binom
import numpy as np

# plane has 450 seats, sells 460 tickets
# Each passenger has 3% chance of missing (97% show up)
n = 460  # tickets sold
p = 0.97  # probability of showing up

# we want probability that 450 or fewer show up
prob_all_seated = binom.cdf(450, n, p)

print(f"Challenge 1: Ironhack Airlines")
print(f"Probability all passengers have seats: {prob_all_seated:.4f}")
print(f"That's about {prob_all_seated*100:.2f}%")
```

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
from scipy.stats import geom

# probability of success on first attempt = 0.3
p = 0.3

# need at least 3 attempts means first 2 fail
# P(X >= 3) = 1 - P(X < 3) = 1 - P(X=1) - P(X=2)

prob_1 = geom.pmf(1, p)
prob_2 = geom.pmf(2, p)

prob_at_least_3 = 1 - prob_1 - prob_2

print("Challenge 2: Call Center")
print(f"Probability needs at least 3 attempts: {prob_at_least_3:.4f}")
print(f"Or about {prob_at_least_3*100:.2f}%")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
from scipy.stats import poisson

# average 500 visits per hour
# server handles up to 550
lambda_rate = 500

# probability of MORE than 550 visits
prob_overwhelmed = 1 - poisson.cdf(550, lambda_rate)

print("Challenge 3: Website")
print(f"Probability server overwhelmed in 1 hour: {prob_overwhelmed:.4f}")#code here
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
# whats probability not overwhelmed in one hour
prob_not_overwhelmed = 1 - prob_overwhelmed

# probability of being not overwhelmed for all 24 hours
prob_never_overwhelmed_24h = prob_not_overwhelmed ** 24

# probability overwhelmed at least once
prob_overwhelmed_24h = 1 - prob_never_overwhelmed_24h

print(f"Probability of being overwhelmed during the 24 hours: {prob_overwhelmed_24h:.4f}")
print(f"result is {prob_overwhelmed_24h*100:.2f}%")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
from scipy.stats import expon

# average customer every ten mins
# so lambda = 1/10 = 0.1 per minute
mean_time = 10
lambda_rate = 1/mean_time

# probability next customer arrives within 5 minutes
prob_within_5 = expon.cdf(5, scale=mean_time)

print("Challenge 4: Helpdesk")
print(f"Probability customer arrives within 5 min: {prob_within_5:.4f}")
print(f"roughly {prob_within_5*100:.1f}%")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
# probability no customer for 15 minutes (employee gets break)

# this is P(X > 15)
prob_no_customer_15 = 1 - expon.cdf(15, scale=mean_time)

print(f"\nProbability employee gets a break: {prob_no_customer_15:.4f}")
print(f"Result is {prob_no_customer_15*100:.2f}%")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
from scipy.stats import norm

# mean is 150g, std dev is 10g
mean = 150
std = 10

# probability of between 140 and 160
prob_between = norm.cdf(160, mean, std) - norm.cdf(140, mean, std)

print("Challenge 5: Bird Weights")
print(f"Probability weight is between 140-160g: {prob_between:.4f}")
print(f"Final result is {prob_between*100:.2f}%")
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
from scipy.stats import expon

# mean lifetime is 50 hours
mean_lifetime = 50

# probability of it fails within first 30 hours
prob_fail_30 = expon.cdf(30, scale=mean_lifetime)

print("Challenge 6: Electronic Component")
print(f"Probability component fails within 30 hours: {prob_fail_30:.4f}")
print(f"About {prob_fail_30*100:.1f}%")
```
